```python
#SQL_SOURCE = 'https://fb55.carto.com/api/v2/sql?q='
SQL_SOURCE = 'https://serv.cusp.nyu.edu/~hvo/files/citibike.csv'

import urllib2
import urllib
import StringIO
import ast
import pandas as pd

def queryCartoDB(query, format='CSV', source=SQL_SOURCE):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    data = urllib.urlencode({'format': format, 'q': query})
    try:
        response = urllib2.urlopen(source, data)
    except urllib2.HTTPError, e:
        raise ValueError('\n'.join(ast.literal_eval(e.readline())['error']))
    except Exception:
        raise
    return response.read()
```

```python
try:
    import urllib as ulib
except ImportError: #added new except#
    import urllib2 as ulib
except ImportError:
    import urllib3 as ulib
    
try:
    from StringIO import BytesIO as io
except ImportError:
    from io import BytesIO as io

try:
    from urllib import urlencode as urlencode
except ImportError:
    from urllib.parse import urlencode as urlencode
    
try:
    from urllib import urlopen as urlopen
except ImportError:
    from urllib.request import urlopen as urlopen
    
try:
    from urllib2 import HTTPError as HTTPError
except ImportError:
    from urllib.error import HTTPError as HTTPError

import ast
import pandas as pd

API_URL = 'https://fb55.carto.com:443/api/v2/sql?q='
#API_URL = 'https://serv.cusp.nyu.edu/~hvo/files/citibike.csv'

def queryCartoDB(query, formatting = 'CSV', source = API_URL):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    
    data = urlencode({'format': formatting, 'q': query}).encode("utf-8")
    try:
        response = urlopen(source, data)
        return response.read()
    except HTTPError as e:
        raise (ValueError('\n'.join(ast.literal_eval(e.readline())['error'])))
        
def get_data(query):
    try:
        return pd.read_csv(io(queryCartoDB(query)), sep = ',')
    except ValueError as v:
        print (str(v))
```

#### Task 1
Task 1 — Familiarize with SQL Clauses <br>
• Sort data by start_station_id, tripduration <br>
-Only checking trips with duration <= 3 hours 

```python
## 60*60*3 = 10800
query = '''
SELECT start_station_id,tripduration
FROM citibike
WHERE tripduration <=10800
ORDER BY start_station_id,tripduration
;
'''
table = get_data(query)
table
```

## Only show the top/last 10 records (aka head and tail in SQL)

```python
query = '''
SELECT *
FROM citibike
order by cartodb_id
limit 10
;
'''
table = get_data(query)
table
```

```python
query = '''
SELECT *
FROM citibike
order by cartodb_id DESC
limit 10
;
'''
table = get_data(query)
table
```

# • List all unique start_station_id values

```python
query = '''
SELECT distinct start_station_id
FROM citibike
;
'''
table = get_data(query)
table
```

• Aggregation functions:<br> 
- Count the number of trips (aka wc -l in SQL) <br>
- Find the average/min/max trip duration <br> 

```python
query = '''
SELECT count(cartodb_id) as num_trips, avg(tripduration) as average_trip,min(tripduration) as min_trip,max(tripduration) as max_trip
FROM citibike
;
'''
table = get_data(query)
table
```

# Task 2

## Selecting trips started on Feb-02-2015 only

```python
query = '''
SELECT *
FROM citibike
where CAST(starttime AS DATE)= DATE'2015-02-02'
;
'''
table = get_data(query)
table
```

## Selecting trips started on the weekends 
### What are average trip duration during weekends

```python
query = '''
SELECT avg(tripduration) as average_weekend_trips
FROM citibike
where EXTRACT(DOW FROM starttime) in (0,6)
;
'''
table = get_data(query)
table
```

## Can we do the same for weekday?

```python
query = '''
SELECT avg(tripduration) as average_weekdays_trips
FROM citibike
where EXTRACT(DOW FROM starttime) not in (0,6)
;
'''
table = get_data(query)
table
```

# Task 3 — Working with Space

• Showing the list of start station locations <br>
-Using GROUP BY 

```python
query = '''
SELECT start_station_name
FROM citibike
group by start_station_name
;
'''
table = get_data(query)
table
```

## Showing the number of trips started per station

```python
query = '''
SELECT start_station_name,count(cartodb_id) as num_trips
FROM citibike
group by start_station_name
;
'''
table = get_data(query)
table
```

... but only for stations within 500m of Time Square! 
- The coordinates of Time Square is (40.7577,-73.9857)

```python
query = '''
select start_station_name,count(cartodb_id) as num_trips
FROM citibike
where ROUND(6378.138 * 2 * ASIN(SQRT(POW(SIN((40.7577 * PI() / 180 - start_station_latitude * PI() / 180) / 2),2) 
+ COS(40.7577 * PI() / 180) * COS(start_station_latitude * PI() / 180) * 
POW(SIN((-73.9857 * PI() / 180 - start_station_longitude * PI() / 180) / 2),2))) * 1000)
<500
group by start_station_name
;
'''
table = get_data(query)
table
```

## Task 4 — Putting it all together

## Find the station that had the longest average trip duration during weekends and within 500m of TimeSquare! 

```python
query = '''
select start_station_name,avg(tripduration) as avg_trip_duration
FROM citibike
where ROUND(6378.138 * 2 * ASIN(SQRT(POW(SIN((40.7577 * PI() / 180 - start_station_latitude * PI() / 180) / 2),2) 
+ COS(40.7577 * PI() / 180) * COS(start_station_latitude * PI() / 180) * 
POW(SIN((-73.9857 * PI() / 180 - start_station_longitude * PI() / 180) / 2),2))) * 1000)
<500 and EXTRACT(DOW FROM starttime) in (0,6) 
group by start_station_name
ORDER BY avg_trip_duration DESC
LIMI
;
'''
table = get_data(query)
table
```
