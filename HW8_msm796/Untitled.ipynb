```python
#from multiprocessing.pool import ThreadPool as Pool
from multiprocessing import Pool
from random import randint
from time import sleep


def process_line(l):
    print(l, "started")
    sleep(randint(0,3))
    print(l, "done")


def get_next_line():
    with open("/nfshome/msm796/PUIdata/yellow_tripdata_2018-06.csv",'r') as f:
        for line in f:
            yield line

f = get_next_line()

t = Pool(processes=8)

for i in f:
    t.map(process_line, (i,))

t.join()
t.close()
```
